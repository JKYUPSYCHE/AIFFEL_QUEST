```python
y_proba = model.predict_proba(X_test)[:, 1]  # ëª¨ë¸ìì ìì¸¡í ë°í(1) íë¥ 
y_proba.mean()
```

* ëª¨ë¸ì ë°ì´í°ì í´ëì¤ ë¹ì¨ì íìµíê³  ììì ì ì ìë¤.

```python
# ì¤ì  ë°í ë¹ì¨ì ë°ìíì¬ ìê³ê° íë³´ ì¤ì 
base_rate = y_test.mean()

thresholds = [
    base_rate,
    base_rate * 1.5,
    base_rate * 2,
    base_rate * 3,
    0.05
]

thresholds = [round(t, 4) for t in thresholds]
thresholds
```

```python
# ìê³ê° íë³´ë¤ ì¤ ê°ì¥ ëì F-1 scoreì ìê³ê° ì±í
results = []
for t in thresholds:
    y_pred_t = (y_proba >= t).astype(int)

    results.append({
        'Threshold': t,
        'Precision': precision_score(y_test, y_pred_t),
        'Recall': recall_score(y_test, y_pred_t),
        'F1-score': f1_score(y_test, y_pred_t)
    })

threshold_df = pd.DataFrame(results)
threshold_df
```

```python
best_row = threshold_df.loc[threshold_df['F1-score'].idxmax()]
best_threshold = best_row['Threshold']
best_threshold
```

```python
y_pred_best = (y_proba >= best_threshold).astype(int)

cm_best = confusion_matrix(y_test, y_pred_best)
cm_df = pd.DataFrame(
    cm_best,
    index=['Actual: No Return', 'Actual: Return'],
    columns=['Predicted: No Return', 'Predicted: Return']
)

cm_df
```
